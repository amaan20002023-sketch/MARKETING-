#### 2. Performing Customers/sellers Segmentation 
a. Divide the customers into groups based on the revenue generated  
b. Divide the sellers into groups based on the revenue generated  

```python
customer_rev=master_table[master_table['order_status']=='delivered']
```

```python
customer_rev=pd.DataFrame(customer_rev.groupby('customer_unique_id')['Revenue'].sum()).reset_index()
customer_rev=customer_rev.sort_values('Revenue',ascending=False)
customer_rev['group']=pd.qcut(customer_rev['Revenue'].rank(method='first'),[0,0.5,0.8,1],labels=['low','medium','high'])
```

```python
customer_rev.head()
```

```python
pd.DataFrame(customer_rev.groupby('group')['Revenue'].sum()).plot(kind='bar')
plt.title('category wise spending')
plt.ylabel('revenue in million')
```

b. Divide the sellers into groups based on the revenue generated

```python
seller_rev=pd.DataFrame(master_table[master_table['order_status']=='delivered'].groupby('seller_id')\
                                ['Revenue'].sum())  
seller_rev=seller_rev.sort_values('Revenue',ascending=False)
seller_rev.head()
```

```python
seller_rev['group']=pd.qcut(seller_rev['Revenue'].rank(method='first'),[0,0.5,0.8,1],labels=['low','medium','high'])
```

```python
seller_rev.head()
```

```python
seller_rev.groupby('group')['Revenue'].sum().plot(kind='bar')
plt.title('category wise seller')
plt.ylabel('sale in million')
```
